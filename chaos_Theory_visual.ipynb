```python
import matplotlib.pyplot as plt
import pandas as pd

def lorenz_attractor(sigma, rho, beta, steps=5000, dt=0.01):
    x, y, z = 0.01, 0.01, 0.01

    xs, ys, zs = [], [], []

    for _ in range(steps):
        dx = sigma * (y - x)
        dy = x * (rho - z) - y
        dz = x * y - beta * z

        x += dx * dt
        y += dy * dt
        z += dz * dt

        xs.append(x)
        ys.append(y)
        zs.append(z)

    return xs, ys, zs


xs, ys, zs = lorenz_attractor(
    sigma=10,
    rho=28,
    beta=8/3
)

fig = plt.figure()
ax = fig.add_subplot(projection="3d")
ax.plot(xs, ys, zs, lw=0.5)

plt.show()

data_rows = [[xs],[ys],[zs]]
coloumn_names = ['X-Value', 'Y-value', 'Z-value']

print(pd.DataFrame(data_rows, coloumn_names))
```
